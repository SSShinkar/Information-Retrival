```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Load and prepare the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
data
```

```python
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
scaled_data
```

```python
#  Apply Agglomerative Clustering
import warnings
warnings.filterwarnings('ignore')
agg_cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
model = agg_cluster.fit_predict(scaled_data)
```

```python
model
```

```python
# Cell 5: Visualize the clusters (using the first two features for simplicity)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=scaled_data[:, 0], y=scaled_data[:, 1], hue=model, palette='viridis', s=60)
plt.title('Agglomerative Clustering on Iris Dataset')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```
